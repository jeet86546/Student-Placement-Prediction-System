```python
import joblib
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
model = joblib.load("placement_model.pkl")
```

```python
print("\nEnter student details for placement prediction:\n")

branch = input("Branch (CSE/ECE/ME/IT): ").strip().upper()
cgpa = float(input("CGPA: "))
tenth_percentage = float(input("10th Percentage: "))
twelfth_percentage = float(input("12th Percentage: "))
backlogs = int(input("Number of Backlogs: "))
attendance_percentage = float(input("Attendance Percentage: "))
projects_completed = int(input("Projects Completed: "))
internships_completed = int(input("Internships Completed: "))
coding_skill_rating = int(input("Coding Skill Rating (1-10): "))
communication_skill_rating = int(input("Communication Skill Rating (1-10): "))
aptitude_skill_rating = int(input("Aptitude Skill Rating (1-10): "))
hackathons_participated = int(input("Hackathons Participated: "))
certifications_count = int(input("Certifications Count: "))
city_tier = input("City Tier (1/2/3): ")
extracurricular_involvement = input("Extracurricular Involvement (0 = No, 1 = Yes): ")
```

```python
numerical_feature =  [cgpa,tenth_percentage,twelfth_percentage,backlogs,
                      attendance_percentage,projects_completed,internships_completed,
                      coding_skill_rating,communication_skill_rating,aptitude_skill_rating,
                      hackathons_participated,certifications_count]
categorical_feature = [branch,city_tier,extracurricular_involvement]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
    ('cat', OneHotEncoder(handle_unknown='ignore'),categorical_feature),
    ('num', 'passthrough',numerical_feature)
    ]
    
)
```

```python
x = pd.DataFrame([{
'branch': branch,
'cgpa': cgpa,
'tenth_percentage': tenth_percentage,
'twelfth_percentage': twelfth_percentage,
'backlogs': backlogs,
'attendance_percentage': attendance_percentage,
'projects_completed': projects_completed,
'internships_completed': internships_completed,
'coding_skill_rating': coding_skill_rating,
'communication_skill_rating': communication_skill_rating,
'aptitude_skill_rating': aptitude_skill_rating,
'hackathons_participated': hackathons_participated,
'certifications_count': certifications_count,
'city_tier': city_tier,
'extracurricular_involvement': extracurricular_involvement
}])
```

```python
y = model.predict(x)
if y== 1:
    print("You might get placed in college Placement")
elif y ==0:
    print("Your chances are very less to get placed, Please do work on your skills.")
```
